## price per sq ft

```python
train_df['price_per_sqft'] = train_df['price'] / train_df['sqft_living']

sns.histplot(train_df['price_per_sqft'], bins=50)
plt.title("Distribution of Price per Sqft")
plt.show()
sns.boxplot(x='waterfront', y='price_per_sqft', data=train_df)
plt.title("Waterfront Premium (Price per Sqft)")
plt.show()
```

## heatmap for correlation

```python
corr_cols = [
    'price', 'sqft_living', 'grade',
    'bathrooms', 'sqft_living15',
    'view', 'waterfront'
]

sns.heatmap(
    train_df[corr_cols].corr(),
    annot=True,
    cmap='coolwarm',
    fmt=".2f"
)
plt.title("Feature Correlation with Price")
plt.show()
```
